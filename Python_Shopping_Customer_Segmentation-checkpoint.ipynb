# Clustering - Univariate, Biavariate, Multivariate

```python
# Initiate the algorithm

clustering1 = KMeans(n_clusters=6)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
# How many of ours customers falls in each cluster
df['Income Cluster'].value_counts()
```

```python
# Find the ideal amount of clusters that we need

clustering1.inertia_ # represent the distance between centroids
```

```python
# We need to use a different number of clusters

# we need to create a efficient way to check the inertia_ on each one of the clusters

inertia_scores = []

for i in range(1,11):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
                   
```

```python
inertia_scores
```

```python
plt.plot(range(1,11), inertia_scores);
```

We are using the elbow method. So, as we can see in this plot. the elbow begins in 3.
We going to re-initiate our algorithm with n_clusters = 3.

```python
# Initiate the algorithm

clustering1 = KMeans(n_clusters=3)
```

```python
clustering1.fit(df[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
df['Income Cluster'] = clustering1.labels_
df.head()
```

```python
df['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
inertia_scores = []

for i in range(1,11):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
```

### Biavariate Clustering

```python
# Initiate the algorithm

clustering2 = KMeans(n_clusters=5)
clustering2.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
df['Spending and Income Cluster'] = clustering2.labels_ 
df.head()
```

```python
inertia_scores2 = []

for i in range(1,11):
    kmeans2 = KMeans(n_clusters= i)
    kmeans2.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
    inertia_scores2.append(kmeans2.inertia_)

plt.plot(range(1,11), inertia_scores2);
```

The elbow here is more around number 5, which is the number of cluster that we will use.

```python
centers = pd.DataFrame(clustering2.cluster_centers_)
centers.columns = ['x', 'y']
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x= centers['x'], y= centers['y'], s= 100, c= 'black', marker= '*')
sns.scatterplot(data=df, x= 'Annual Income (k$)', y= 'Spending Score (1-100)', 
                hue= 'Spending and Income Cluster', palette= 'tab10');
plt.savefig('clustering_bivariate.png')
```

```python
pd.crosstab(df['Spending and Income Cluster'], df['Gender'])
```

```python
pd.crosstab(df['Spending and Income Cluster'], df['Gender'], normalize= 'index')
```

### Multivariate Clustering

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale= StandardScaler()
```

```python
# converting the columns Gender into numbers:

dff = pd.get_dummies(df, drop_first= True)
dff.head()
```

```python
dff.columns
```

```python
dff.rename(columns={'Gender_Male': 'Gender'}, inplace= True)
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Gender']]
dff.head()
```

```python
dff = scale.fit_transform(dff)
```

```python
dff = pd.DataFrame(scale.fit_transform(dff))
dff.head()
```

```python
inertia_scores3 = []

for i in range(1,11):
    kmeans3 = KMeans(n_clusters= i)
    kmeans3.fit(dff)
    inertia_scores3.append(kmeans3.inertia_)

plt.plot(range(1,11), inertia_scores3);
```
